```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/kaggle/input/position/Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

# **TRAINING THE POLYNOMIAL REGRESSION MODEL ON THE WHOLE DATASET**

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg  = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
plt.scatter(X,y,color='orange')
plt.plot(X,lin_reg.predict(X), color = 'pink')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,y,color='orange')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(X, y)
print(mae)
```

**Predicting the result in linear regression**

```python
lin_reg.predict([[7]])
```

**Predicting the result in polynomial regression**

```python
lin_reg_2.predict(poly_reg.fit_transform([[7]]))
```
